In Statistical Mechanics the one-dimensional (1D) Ising Model is exactly resoluble. The Hamiltonian for a system of N spins with nearest neighbors interaction is given by (in the following I will assume units such that $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model),and with the assumption $s_{N+1} = s_{1}$.

The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:

$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.
Once the partition function has been computed, the Helmholtz free energy is readly obtained as $A = -k_B T \ln Z$. All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$:

- the **Internal Energy** per spin (with $h=0$):

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

- the **Heat Capacity** (with $h=0$):

$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= \frac{\partial \beta}{\partial T}\frac{\partial U}{\partial \beta}
= -k_B\beta^2 \frac{\partial}{\partial \beta} \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z} \right] = \\
= -k_B\beta^2 \left[ -\sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \sum_{\{ s_i \}} H \exp \left( -\beta H\right)\frac{\partial Z^{-1}}{\partial \beta} \right] = \\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]\frac{1}{Z}\frac{\partial Z}{\partial \beta} \right\} =\\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} - \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]^2 \right\} =
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

- The **magnetization** (with $h \not= 0$):

Note that with $h=0$ the magnetization $M(N,T,h=0)=\langle \sum_{i=1,N}s_i \rangle =0$ because in 1D we cannot have a (ferromagnetic) phase transition in absence of an external magnetic field.

$$
M(N,T,h) = -\frac{\partial A}{\partial h} = k_B T \frac{\partial \ln Z}{\partial h} =
\frac{1}{\beta} \frac{1}{Z}\frac{\partial}{\partial h}\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} = \left\langle \sum_{i=1,N}s_i \right\rangle
$$

The exact formula from $M(N,T,h) = k_B T \dfrac{\partial \ln Z}{\partial h}$ is:
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

- The **magnetic susceptibility** (with $h=0$):

$$
\chi(N,T) = \frac{\partial M}{\partial h} = 
\frac{\partial}{\partial h}\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} =\\
= \beta\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right]^2 \frac{\exp \left( -\beta H\right)}{Z}
-\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z}
\frac{1}{Z}\frac{\partial Z}{\partial h} = 
\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

### 1) Simulazione del modello di Ising con algoritmo di Metropolis

L'obiettivo di questo esercizio è applicare l'algoritmo di Metropolis al modello di Ising 1D, un modello fondamentale in fisica statistica. Il sistema è costituito da spin $s_i = \pm 1$ disposti su una catena unidimensionale di lunghezza $N$, con interazioni tra spin limitate ai primi vicini e una possibile interazione con un campo magnetico esterno. La Hamiltoniana del sistema è definita come:

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

dove $J$ rappresenta l'accoppiamento tra spin e $h$ il campo magnetico esterno.

L'algoritmo di Metropolis per questo modello funziona nel seguente modo:
- Si parte da una configurazione di spin iniziale $\{s_i\}$.
- Si genera una nuova configurazione $\{s'_i\}$ flippando casualmente uno spin.
- Si calcolano le energie associate $E$ e $E'$ delle configurazioni $\{s_i\}$ e $\{s'_i\}$ rispettivamente.
- Si accetta la nuova configurazione $\{s'_i\}$ con probabilità $p(\{s_i\} \rightarrow \{s'_i\}) = \min\left[1, e^{-\beta (E' - E)}\right]$, dove $\beta = \frac{1}{k_B T}$ con $k_B$ costante di Boltzmann e $T$ temperatura.

In questa simulazione, sono stati utilizzati $N = 50$ spin con condizioni periodiche al contorno. Il sistema è stato raffreddato da $T = 3$ a $T = 0.5$ con decrementi di $0.1$. Prima di ogni misura, sono stati eseguiti $5 \times 10^5$ passi per garantire l'equilibrazione del sistema. Le misure sono state effettuate utilizzando la tecnica delle medie a blocchi con $N_{blocks} = 50$ e $N_{step} = 50000$ passi per blocco.

Per facilitare l'equilibrazione durante il raffreddamento, per ogni nuova temperatura il sistema è stato fatto ripartire dalla configurazione finale della temperatura precedente anziché da una configurazione casuale, diminuendo così gli step necessari per raggiungere il nuovo equilibrio.



```python
############################################################################
# Import librerie e calcali analitici
############################################################################

import numpy as np
import matplotlib.pyplot as plt

# Funzione per calcolare il contributo alla chi² per un singolo punto di dati
def chi2_i(Obs, Exp, err):
    return ((Obs-Exp)/err)**2

# Impostazione dei parametri e calcolo della temperatura
points = 100
T = np.linspace(0.1, 3, num=points)  # Gamma di temperature da 0.1 a 3 con 100 punti
beta = 1 / T  # Inverso della temperatura
J = 1.0  # Costante di accoppiamento
Ns = 50  # Numero di siti nel modello di Ising 1D

# Funzione per calcolare l'energia interna teorica
def calculate_internal_energy(Temp, J, Ns):
    th = np.tanh(J / Temp)  # Iperbolico tangente di J / Temp
    thN = th ** Ns  # th elevato a Ns
    ch = 1 / th  # Iperbolico coseno di J / Temp
    e = -J * (th + ch * thN) / (1 + thN)  # Energia interna
    return e

e = calculate_internal_energy(T, J, Ns)  # Calcolo dell'energia interna teorica

# Funzione per calcolare il calore specifico teorico
def calculate_specific_heat(Temp, J, Ns):
    beta = 1. / Temp  # Inverso della temperatura
    th = np.tanh(J / Temp)  # Iperbolico tangente di J / Temp
    thN = th ** Ns  # th elevato a Ns
    ch = 1 / th  # Iperbolico coseno di J / Temp
    return ((beta * J) ** 2) * (((1 + thN + (Ns - 1) * (th ** 2) + (Ns - 1) * (ch ** 2) * thN) / (1 + thN)) - Ns * ((th + ch * thN) / (1 + thN)) ** 2)

cv = calculate_specific_heat(T, J, Ns)  # Calcolo del calore specifico teorico

# Funzione per calcolare la magnetizzazione teorica
h = 0.02  # Campo esterno
def calculate_magnetization(Temp, J, h, Ns):
    beta = 1. / Temp  # Inverso della temperatura
    l1 = np.exp(beta * J) * np.cosh(beta * h) + np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) * np.cosh(beta * h) - 2 * np.sinh(2 * beta * J))
    l2 = np.exp(beta * J) * np.cosh(beta * h) - np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) * np.cosh(beta * h) - 2 * np.sinh(2 * beta * J))
    Z = l1 ** Ns + l2 ** Ns  # Funzione di partizione
    return (np.exp(beta * J) * np.sinh(beta * h) * ((l1 ** (Ns - 1)) * (1 + np.exp(beta * J) * np.cosh(beta * h) / np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) * np.cosh(beta * h) - 2 * np.sinh(2 * beta * J)))
                                                  + (l2 ** (Ns - 1)) * (1 - np.exp(beta * J) * np.cosh(beta * h) / np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) * np.cosh(beta * h) - 2 * np.sinh(2 * beta * J))))) / (Z)

M = calculate_magnetization(T, J, h, Ns)  # Calcolo della magnetizzazione teorica

# Funzione per calcolare la suscettibilità teorica
def calculate_susceptibility(Temp, J, Ns):
    beta = 1. / Temp  # Inverso della temperatura
    th = np.tanh(J / Temp)  # Iperbolico tangente di J / Temp
    thN = th ** Ns  # th elevato a Ns
    return beta * np.exp(2 * beta * J) * (1 - thN) / (1 + thN)

X = calculate_susceptibility(T, J, Ns)  # Calcolo della suscettibilità teorica
```

```python
###########################################################################
# Metropolis
###########################################################################

# Caricamento dei dati per energia interna
data = np.loadtxt('OUTPUT_MNT/total_energy.dat')
temperatura = data[:, 0]
energia = data[:, 1]
errore_energia = data[:, 2]

# Caricamento dei dati per il calore specifico
data2 = np.loadtxt('OUTPUT_MNT/specific_heat.dat')
heat_capacity = data2[:, 1]
errore_heat = data2[:, 2]

# Caricamento dei dati per la magnetizzazione
data3 = np.loadtxt('OUTPUT_MNT/magnetization.dat')
magn = data3[:, 1]
errore_magn = data3[:, 2]

# Caricamento dei dati per la suscettibilità
data4 = np.loadtxt('OUTPUT_MNT/susceptibility.dat')
susc = data4[:, 1]
errore_susc = data4[:, 2]

# Creazione dei plot
fig1, ((ax1, ax2), (ax5, ax6)) = plt.subplots(2, 2, figsize=(20, 8), gridspec_kw={'height_ratios': [10, 5]}, sharex=True)
fig1.suptitle('Simulazione modello di Ising con Metropolis', fontsize=30)
plt.subplots_adjust(hspace=0)
fig2, ((ax3, ax4), (ax7, ax8)) = plt.subplots(2, 2, figsize=(20, 8), gridspec_kw={'height_ratios': [10, 5]}, sharex=True)
fig2.suptitle('Simulazione modello di Ising con Metropolis', fontsize=30)
plt.subplots_adjust(hspace=0)


def symmetrize_y_axis(axes):
    y_max = np.abs(axes.get_ylim()).max()
    axes.set_ylim(ymin=2-y_max, ymax=y_max)

# Plot del primo grafico: Energia interna
energia_teor = calculate_internal_energy(temperatura, J, Ns)
chi2_energia = sum(chi2_i(energia, energia_teor, errore_energia))/len(energia)
ax1.plot(T, e, label='Teoria')
ax1.errorbar(temperatura, energia, yerr=errore_energia, fmt='o', label='Simulazione')
ax1.set_title('Energia Interna', fontsize=18)
ax1.set_ylabel(r'Energia interna $[U/JN]$', fontsize=15)
ax1.legend()
ax1.text(0.05, 0.80, r'$\chi_r^2$= {:.2f}'.format(chi2_energia), ha='left', va='top', transform=ax1.transAxes, bbox=dict(facecolor='white', alpha=1, pad=8), fontsize=15)
ax1.grid()

# Plot del secondo grafico: Calore specifico
heat_capacity_teorico = calculate_specific_heat(temperatura, J, Ns)
chi2_heat_capacity = sum(chi2_i(heat_capacity, heat_capacity_teorico, errore_heat)) / len(heat_capacity)
ax2.plot(T, cv, label='Teoria')
ax2.errorbar(temperatura, heat_capacity, yerr=errore_heat, fmt='o', label='Simulazione')
ax2.set_title('Calore Specifico', fontsize=18)
ax2.set_ylabel(r'Calore specifico $[C/k_B N]$', fontsize=15)
ax2.legend()
ax2.text(0.76, 0.80, r'$\chi_r^2$= {:.2f}'.format(chi2_heat_capacity), ha='left', va='top', transform=ax2.transAxes, bbox=dict(facecolor='white', alpha=1, pad=8), fontsize=15)
ax2.grid()

# Plot del terzo grafico: Magnetizzazione
magn_teorico = calculate_magnetization(temperatura, J, h, Ns)
chi2_magnetization = sum(chi2_i(magn, magn_teorico, errore_magn)) / len(magn)
ax3.plot(T, M, label='Teoria')
ax3.errorbar(temperatura, magn, yerr=errore_magn, fmt='o', label='Simulazione')
ax3.set_title('Magnetizzazione (h=0.02)', fontsize=18)
ax3.set_ylabel('Magnetizzazione $[M/N]$', fontsize=15)
ax3.legend()
ax3.text(0.76, 0.80, r'$\chi_r^2$= {:.2f}'.format(chi2_magnetization), ha='left', va='top', transform=ax3.transAxes, bbox=dict(facecolor='white', alpha=1, pad=8), fontsize=15)
ax3.grid()

# Plot del quarto grafico: Suscettibilità
susc_teorico = calculate_susceptibility(temperatura, J, Ns)
chi2_susc = sum(chi2_i(susc, susc_teorico, errore_susc)) / len(susc)
ax4.plot(T, X, label='Teoria')
ax4.errorbar(temperatura, susc, yerr=errore_susc, fmt='o', label='Simulazione')
ax4.set_title('Suscettività', fontsize=18)
ax4.set_ylabel(r'Suscettività $[\chi]$', fontsize=15)
ax4.legend()
ax4.text(0.76, 0.80, r'$\chi_r^2$= {:.2f}'.format(chi2_susc), ha='left', va='top', transform=ax4.transAxes, bbox=dict(facecolor='white', alpha=1, pad=8), fontsize=15)
ax4.grid()

# Plot del rapporto tra teoria e simulazione per l'energia interna
ax5.errorbar(temperatura, energia / energia_teor, yerr=abs(errore_energia/energia_teor), fmt='o', label='Rapporto tra punti simulati e teorici')
ax5.set_xlabel(r'Temperatura $[k_B T/J]$', fontsize=15)
ax5.set_ylabel('$U_{sim}/U_{teo}$', fontsize=12)
ax5.set_ylim(0.998,1.002)
ax5.axhline(y=1, color='r', linestyle='--')
ax5.legend()
ax5.grid()

# Plot del rapporto tra teoria e simulazione per il calore specifico
ax6.errorbar(temperatura, heat_capacity / heat_capacity_teorico, yerr=errore_heat/heat_capacity_teorico, fmt='o', label='Rapporto tra punti simulati e teorici')
ax6.set_xlabel(r'Temperatura $[k_B T/J]$', fontsize=15)
ax6.set_ylabel('$C_{sim}/C_{teo}$', fontsize=12)
ax6.set_ylim(0.98,1.02)
ax6.axhline(y=1, color='r', linestyle='--')
ax6.legend()
ax6.grid()

# Plot del rapporto tra teoria e simulazione per la magnetizzazione
ax7.errorbar(temperatura, magn / magn_teorico, yerr=errore_magn/magn_teorico, fmt='o', label='Rapporto tra punti simulati e teorici')
ax7.set_xlabel(r'Temperatura $[k_B T/J]$', fontsize=15)
ax7.set_ylabel(r'$M_{sim}/M_{teo}$', fontsize=12)
ax7.set_ylim(0.9,1.1)
ax7.axhline(y=1, color='r', linestyle='--')
ax7.legend()
ax7.grid()

# Plot del rapporto tra teoria e simulazione per la suscettibilità
ax8.errorbar(temperatura, susc / susc_teorico, yerr=errore_susc/susc_teorico, fmt='o', label='Rapporto tra punti simulati e teorici')
ax8.set_xlabel(r'Temperatura $[k_B T/J]$', fontsize=15)
ax8.set_ylabel(r'$\chi_{sim}/\chi_{teo}$', fontsize=12)
ax8.set_ylim(0.98,1.02)
ax8.axhline(y=1, color='r', linestyle='--')
ax8.legend()
ax8.grid()


plt.show()
```

### 2) Simulazione del modello di Ising con il metodo di Gibbs

Un altro algoritmo che si può implementare per simulare questo sistema è quello di Gibbs, che si può vedere come un caso particolare dell'algoritmo di Metropolis ma con la peculiarità e il vantaggio che l'accetazione avviene con probabilità 1 sempre. Funziona nel seguente modo:
- si seleziona casualmente uno spin ad ogni step;
- una volta fissati tutti gli altri, si sceglie un nuovo valore per questo con una probabilità

\begin{equation}
p = p(s_k'|\{s_j:\ j \neq k\})= \frac{p(s_1, ..., s_k', ...,s_N)}{\sum_{s_k = \pm 1} p(s_1, ..., s_k, ..., s_N)}
\end{equation}

Anche in questo caso la convergenza alla soluzione è garantita nel limite di step infiniti.
